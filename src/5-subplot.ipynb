### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Define data for line plot
x_line = np.linspace(0, 10, 100)
y_line = x_line ** 2
axs[0, 0].plot(x_line, y_line, color='blue')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('X values')
axs[0, 0].set_ylabel('Y values')
axs[0, 0].grid(True)

# Define data for bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']
axs[0, 1].bar(categories, values, color=colors)
axs[0, 1].set_title('Bar Plot')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

# Define data for scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x_scatter, y_scatter, color='blue')
for i, j in zip(x_scatter, y_scatter):
    axs[1, 0].text(i, j, f'({i}, {j})', fontsize=10, ha='right')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X values')
axs[1, 0].set_ylabel('Y values')
axs[1, 0].grid(True)

# Define data for pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen', 'pink'], startangle=140)
axs[1, 1].set_title('Pie Chart')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
